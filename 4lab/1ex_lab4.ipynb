1. Сгенерировать случайным образом массив numpy из 1000 значений
нормально распределенной случайной величины (numpy.random.normal) с
мат.
ожиданием
M = 1.0
и стандартным отклонением
s = 1.0.
Преобразовать его в объект Series.

```python
normal_array = np.random.normal(loc=1.0, scale=1.0, size=1000)
normal_series = pd.Series(normal_array)
print(normal_series)
```

2. Вычислить, какая доля всех значений находится в диапазоне (M-s; M+s).

```python
sigma_range = ((normal_series > 0.0) & (normal_series < 2.0)).mean()
print(sigma_range)
```
