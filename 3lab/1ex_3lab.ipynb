Для стран загружаем информацию из файла

```python
countries = np.genfromtxt('global-electricity-generation.csv',
                          dtype= str,
                          delimiter= ',',
                          skip_header=1,
                          usecols=0
                          )
print(countries)
```

Для производства энергии загружаем информацию из файла

```python
generation = np.genfromtxt('global-electricity-generation.csv',
                           delimiter=',',
                           skip_header=1,
                           usecols=range(1, 31)
                           )
print(generation)
```

Для потребления загружаем информацию из файла

```python
consumption = np.genfromtxt('global-electricity-consumption.csv',
                           delimiter=',',
                           skip_header=1,
                           usecols=range(1, 31)
                           )
print(consumption)
```

2) Строим одномерные массивы ежегодного производства и потребления
электроэнергии в среднем за последние 5 лет

```python
average_generation = generation[:, -5:].sum(axis=1)/5
print(average_generation)
```

```python
average_consumption = consumption[:, -5:].sum(axis=1)/5
print(average_consumption)
```

3) Суммарное (по всем странам) потребление электроэнергии за каждый
год.

```python
consumption_sum_for_years = consumption.sum(axis=1)
print(consumption_sum_for_years)
```

3.2. Максимальное количество электроэнергии, которое произвела одна

```python
max_generation = np.nanmax(generation, axis=1)
print(max_generation)
```

3.3. Список стран, которые производят более 500 млрд. кВт*ч
электроэнергии ежегодно в среднем за последние 5 лет

```python
countries_with_big_generation = []
for i in range(len(average_generation)):
  if average_generation[i] > 500:
    countries_with_big_generation.append(countries[i])
print(", ".join(countries_with_big_generation))
```

3.4. 10% стран, которые потребляют больше всего электроэнергии
ежегодно в среднем за последние 5 лет

```python
quantile_90 = np.quantile(average_consumption, 0.9)
countries_with_big_consumption = []
for i in range(len(average_consumption)):
  if average_consumption[i] > quantile_90:
    countries_with_big_consumption.append(countries[i])
print(", ".join(countries_with_big_consumption))
```

3.5. Список стран, которые увеличили производство электроэнергии в 2021
году по сравнению с 1992 годом более, чем в 10 раз.

```python
countries_with_increased_generation = []
for i in range(len(countries)):
  first_year = generation[i, 0]
  last_year = generation[i, -1]

  if first_year > 0 and not np.isnan(first_year) and not np.isnan(last_year):
      if (last_year / first_year) > 10:
          countries_with_increased_generation.append(countries[i])

print(", ".join(countries_with_increased_generation))
```

3.6. Список стран, которые в сумме за все годы потратили больше 100 млрд.
кВт*ч электроэнергии и при этом произвели меньше, чем потратили.

```python
unprofitable_countries = []
for i in range(len(countries)):
  consumption_sum = sum(consumption[i,:])
  generation_sum = sum(generation[i,:])

  if consumption_sum > 100 and generation_sum < consumption_sum:
    unprofitable_countries.append(countries[i])

print(", ".join(unprofitable_countries))
```
